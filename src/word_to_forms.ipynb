# Script

```python
from docx import Document
from docx.enum.text import WD_PARAGRAPH_ALIGNMENT
from docx.shared import Pt


def create_bolao_doc(
    text: str,
    rodada: int,
    output_file: str,
):
    document = Document()

    # ==================================================
    # CAPA
    # ==================================================
    document.add_paragraph(
        f"{rodada} Rodada",
        style="Title"
    )

    document.add_paragraph(
        "Primeira Rodada de chutes do bolão da Copa do Mundo 2026!!!",
        style="Subtitle"
    )

    # página exclusiva para o título
    document.add_page_break()


    # ==================================================
    # PARTIDAS
    # ==================================================
    blocks = [
        block.strip()
        for block in text.strip().split("\n\n")
        if block.strip()
    ]

    for idx, block in enumerate(blocks):

        lines = [
            line.strip()
            for line in block.splitlines()
            if line.strip()
        ]

        partida = lines[0]
        time_a = lines[1]
        time_b = lines[2]

        document.add_paragraph(
            partida,
            style="Heading 1"
        )

        document.add_paragraph()

        document.add_paragraph(
            f"  1. {time_a}",
            style="Normal"
        )

        document.add_paragraph(
            f"  2. {time_b}",
            style="Normal"
        )

        if idx < len(blocks) - 1:
            document.add_page_break()

    print("Salvando arquivo...")
    document.save(output_file)
    print("Arquivo salvo!")
```
